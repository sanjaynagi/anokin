# Sampling Protocol for AnoKin

This protocol describes the sampling procedures for collecting *Anopheles funestus* mosquitoes at Lake Kanyaboli, Kenya, using Prokopack aspirators. For detailed information on Prokopack aspirator use, please refer to the [PacMOSSI Standard Operating Procedure](https://pacmossi.org/carbon/assets/2022/05/PacMOSSI-SOP_Prokopack-1.pdf).

## Study Area and Design

### Location
- Study site: Lake Kanyaboli, Siaya County, western Kenya
- Sampling area: ~3km Ã— 3km wedge-shaped area extending from northeastern corner of lake
- Area divided into four alternating bands (quadrants)

### Timeline
- Primary sampling period: 6 days
- Buffer period: 5 additional days if needed
- Collection times: 06:00 - 10:00 daily
- Duration per house: ~25 minutes

## Equipment Required

### Collection Equipment
- Prokopack aspirators (1 per team)
- Spare batteries (12V)
- Collection cups with mesh lids
- Elastic bands for securing cup lids
- Labels for collection cups
- Marker pens
- Field collection bags
- Tablets with ODK software installed
- Tablet chargers and power banks

### Storage Materials
- 1.5ml Eppendorf tubes
- Barcoded plastic storage bags
- Silica gel packets
- Permanent markers
- Forceps for handling specimens
- Laboratory gloves

### General Field Equipment
- Basic tool kit
- Extra batteries
- Power banks/portable chargers
- Local maps
- Head torches with red lens (~680 nm)

## Household Selection Procedure

1. **Preparation**
   - Identify GPS coordinate within assigned quadrant
   - Locate four nearest compounds to the coordinate
   - Focus on primary sleeping structure in each compound
   - Skip minor structures within compounds

2. **Consent Process**
   - Visit with two fieldworkers plus community facilitator
   - Explain study aims and procedures
   - Present consent form and answer questions
   - If consent given, record GPS coordinates in ODK
   - Inform household to avoid repellents/insecticides night before collection

## Collection Protocol

### Before Starting
1. Organize and label collection cups for each sampling location
2. Assemble Prokopack aspirator following SOP guidelines
3. Test aspirator function
4. Prepare data collection forms/tablets

### Indoor Collection Process
1. **Room Preparation**
   - Close all windows and doors before starting
   - Begin at back rooms, working toward front
   - For multi-story houses, start upstairs and work down
   - Follow left-to-right sequence in each room

2. **Collection Method** (15 minutes per house)
   - Hold aspirator within 15cm of surfaces
   - Move at approximately 1 meter per second
   - Follow systematic pattern:
     a. Sample walls vertically, moving clockwise
     b. Sample ceilings horizontally
     c. Check furniture, especially dark areas
     d. Focus on closets, wardrobes, hanging clothes
     e. Collect any flying mosquitoes observed

3. **Sample Handling**
   - Place collection cup lid before turning off aspirator
   - Label cup with:
     - House ID (VillageID-houseID-date)
     - GPS coordinates
     - Collection time
     - Collector name

### Data Recording
Use ODK form to record:
- Collector Name
- Consent status
- Date collected
- House identifier
- GPS coordinates
- Wall type
- Roof type
- Start Time
- End Time

## Sample Processing

1. **Initial Storage**
   - Transfer mosquitoes to labeled Eppendorf tubes
   - Add silica gel for preservation
   - Each tube should contain exactly one mosquito
   - Label both tube and lid with unique ID

2. **Documentation**
   - Record sample metadata in ODK
   - Take photos of collection site if needed
   - Note any unusual circumstances or deviations from protocol

## Quality Control
- Daily data review by project coordinator
- Feedback to teams on sampling coverage
- Monitor collection rates across quadrants
- Adjust sampling effort if needed to ensure even coverage
